```
def leap(y):
  start=2021
  while start <= y:
    if start % 4 == 0 and start % 100 != 0:
        print(start)
    if start % 100 == 0 and start % 400 == 0:
        print(start)
    start += 1
y=int(input("enter a year :"))
leap(y)
```
